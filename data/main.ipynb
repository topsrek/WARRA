```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from decimal import Decimal
```

```python
Beilage1_filename = "../raw_data/extracted_data/csv_files/Beilage_1_combined_tables.csv"
df_1 = pd.read_csv(Beilage1_filename)
df_1
```

```python
def convert_to_euro(column):
    return pd.to_numeric(column).map("€{:,.2f}".format)
```

```python
df_1
# Create multi-index
df_1["Year"] = pd.Period("2023")
df_1 = df_1.set_index(["Year", "LST"])

df_1["FG-Code"] = "ALL"

# If you need to do calculations, keep a numeric version of the columns
df_1['Refundierungen_pretty'] = convert_to_euro(df_1['Refundierungen'])
df_1['Rechnungsbeträge_pretty'] = convert_to_euro(df_1['Rechnungsbeträge'])


print("Dataset Shape:", df_1.shape)
df_1.head()
```

```python
Beilage2_filename = "../raw_data/extracted_data/csv_files/Beilage_2_combined_tables.csv"
df_2 = pd.read_csv(Beilage2_filename)
df_2
```

```python
# Create multi-index
df_2["Year"] = pd.Period("2023")
df_2 = df_2.set_index(["Year", "FG-Code"])

df_2["LST"] = "ALL"
```

```python
# Display both DataFrames
print("LST Data Shape:", df_1.shape)
print("\nLST Data Sample:")
print(df_1.head())

print("\nFG-Code Data Shape:", df_2.shape)
print("\nFG-Code Data Sample:")
print(df_2.head(3))
```

```python
df_combined = pd.concat([df_1, df_2])
df_combined
```

```python
# Verify LST data
lst_sum = df_combined[
    df_combined.index.get_level_values(1).isin(
        [
            "ÖGK-W",
            "ÖGK-N",
            "ÖGK-B",
            "ÖGK-O",
            "ÖGK-ST",
            "ÖGK-K",
            "ÖGK-S",
            "ÖGK-T",
            "ÖGK-V",
        ]
    )
]["Refundierungen"].sum()
lst_gesamt = df_combined[df_combined.index.get_level_values(1) == "Gesamt"][
    "Refundierungen"
].iloc[0]

# Verify FG-Code data
fg_sum = df_combined[
    df_combined.index.get_level_values(1).astype(str).str.contains(r"^\d+\.0$")
]["Refundierungen"].sum()
fg_gesamt = df_combined[df_combined.index.get_level_values(1).isna()][
    "Refundierungen"
].iloc[0]

print("LST Sum:", lst_sum)
print("LST Gesamt:", lst_gesamt)
print(
    "LST matches:", abs(lst_sum - lst_gesamt) < 1
)  # Using small threshold due to floating point

print("\nFG-Code Sum:", fg_sum)
print("FG-Code Gesamt:", fg_gesamt)
print("FG-Code matches:", abs(fg_sum - fg_gesamt) < 1)

print("\nBoth Gesamt match:", abs(lst_gesamt - fg_gesamt) < 1)
```
